```python
import pandas as pd
weather = pd.DataFrame({
    'South East':['Enugu' ,'Anambra','Imo' ,'Abia' ,'Eboni'],
    'Tempreture': [23,26,20 ,27,19],
    'Humidity' :[78,80,55 ,82,67]
})
weather
```

```python
weather2 = pd.DataFrame({
    'South West':['Lagos' ,'Ogun','Osun' ,'Ekiti' ,'Oyo'],
    'Tempreture': [20,26,10 ,17,12],
    'Humidity' :[88,89,59 ,67,72]
})
weather2
```

```python
weather3 =pd.concat([weather ,weather2])
weather3
```

```python
weather3 =pd.concat([weather ,weather2],sort=False)
weather3
```

```python
weather3 =pd.concat([weather ,weather2] ,sort=True)
weather3
```

# continuing with indexing

```python
weather4 =pd.concat([weather ,weather2] ,ignore_index =True)
weather4
```

```python
weather4 =pd.concat([weather ,weather2] ,ignore_index =True ,keys=['South East' ,'South West'])
weather4
```

```python
weather4 =pd.concat([weather ,weather2] ,keys=['South East' ,'South West'])
weather4
```
