```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_val =[1400 ,600 ,300,410 ,250]
exp_labels =['Home Rent' ,'Food' ,'Telephone Bills' ,'Car' ,'Other Utilities']
```

```python
plt.pie(exp_val ,labels =exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_val ,labels =exp_labels)
plt.show()
```

# increasing the size of the cirlce

```python
plt.axis('equal')
plt.pie(exp_val ,labels =exp_labels ,radius=1.5)
plt.show()
```

# adding percentage to the chart

```python
plt.axis('equal')
plt.pie(exp_val ,labels =exp_labels ,radius=1.5 ,autopct ='%0.1f%%')
plt.show()
```

# bringing out a portion 

```python
plt.axis('equal')
plt.pie(exp_val ,labels =exp_labels ,radius=1.5 ,shadow=True ,explode =[0,0.3,0,0,0])
plt.show()
```
